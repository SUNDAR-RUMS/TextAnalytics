## Webscraping Wiki Links

Task:
1. Read the file Assignment_2_Links.csv. This file contains URL Link and Category
2. Use BeautifulSoup, Scarpe the URL Link provided in Assignment_2_Links.csv
3. Extract all the "Paragraph Tag" from URL
4. Put the paragraph content and category in DICTIONARY
5. Write the results information in results-text-clssification.csv


```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import time
import csv

try:
    # Open the results-text-clssification.csv in WRITE Mode
    csv_file = open('results-text-clssification.csv', 'w', encoding="UTF-8", newline="")
    writer = csv.writer(csv_file)
    
    # Create the Column Name
    writer.writerow(['Text','Category'])
    # Read Assignment_2_Links.csv which contains the URL Links & Category Column
    with open('Assignment_2_Links.csv', mode='r') as csv_input_file:
        csv_reader = csv.DictReader(csv_input_file)

        record_counter = 0
        for row in csv_reader:
            target=row['Category']
            sourceURL=str(row['SOURCEURL'])
            print(target,sourceURL)

            ##############################
            try:
                pageTitleString=""
                htmlPage=""

                htmlLink = sourceURL
                try:
                    urlLink = urlopen(htmlLink)
                    time.sleep(3)
                    htmlPage = urlLink.read()
                    urlLink.close()
                except IOError:
                    print ("Error while opening ".format(htmlLink))
                except Exception as e:
                    print ("Unexcepted Invalid URL error.".format(htmlLink))
                    print ('Reason: ', e)

                htmlPagecontent=""
                contentDetails = {}
                htmlPagecontent = BeautifulSoup(htmlPage)

                # title of the page
                print(htmlPagecontent.title.string)
                pageTitleString=htmlPagecontent.title.string

                paraGraphCounter=0;
                for paragraph in htmlPagecontent.find_all('p'):
                    paraGraphCounter=paraGraphCounter+1
                    #print("Paragraph " + str(paraGraphCounter))
                    paraGraphCounterText=str(paraGraphCounter)
                    #print(str(paragraph.text))
                    paraGraphContent=str(paragraph.text)
                    #paraGraphContent=paraGraphContent.strip()

                    contentDetails['Text'] = paraGraphContent
                    contentDetails['Category'] = target
                    writer.writerow(contentDetails.values())

            except Exception as err:
                print(err)       

            ##############################

            record_counter += 1
        print('Processed  lines from File ' + str(record_counter))
        csv_file.close()
except Exception as e:
    print(e)
    csv_file.close()
print("Completed")

csv_file.close()
```
